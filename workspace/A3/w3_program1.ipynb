```python
import numpy as np
from scipy.signal import triang
from scipy.fftpack import fft
import matplotlib.pyplot as plt

x = triang(15)
X = fft(x)
mX = abs(X)
pX = np.angle(X)
```

```python
plt.plot(mX)
plt.show()
```

```python
plt.plot(pX)
plt.show()
```
